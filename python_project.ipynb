```python
# Chart - 5 visualization code
# Question 5: What is the average price of each room type?
avg_price_room_type = airbnb_df.groupby('room_type')['price'].mean().reset_index()
sns.barplot(x='room_type', y='price', data=avg_price_room_type)
plt.show()
```

```python
# Chart - 9 visualization code
#18. How many listings are available for each price range?

#define the price ranges
price_ranges = ['0-50', '50-100', '100-150', '150-200', '200-250', '250-300', '300-350', '350-400', '400-450', '450-500', '500+']

#create a new column with the price range for each listing
airbnb_df['price_range'] = pd.cut(airbnb_df['price'], bins=[0, 50, 100, 150, 200, 250, 300, 350, 400, 450, 500, np.inf], labels=price_ranges)

#count the number of listings in each price range
listings_by_price_range = airbnb_df.groupby('price_range')['id'].count().reset_index(name='count').head(20)


#plot the bar chart
plt.bar(listings_by_price_range['price_range'], listings_by_price_range['count'])
plt.title('Number of Listings Available for Each Price Range')
plt.xlabel('Price Range ($)')
plt.ylabel('Number of Listings')
plt.show()
```

#**6- STACKED BAR CHART**

```python
# Chart - 6 visualization code
# select the top 10 neighborhoods based on the count of listings
top_10_neighborhoods = airbnb_df['neighbourhood'].value_counts().nlargest(10).index.tolist()

# filter the dataframe to include only the top 10 neighborhoods
filtered_df = airbnb_df.loc[airbnb_df['neighbourhood'].isin(top_10_neighborhoods)]

# group the data by neighborhood and room type to calculate the percentage
room_type_percentages = filtered_df.groupby(['neighbourhood', 'room_type']).size().reset_index(name='count')
room_type_percentages['percentage'] = room_type_percentages.groupby('neighbourhood')['count'].apply(lambda x: x / x.sum() * 100)

# use pivot to reshape the data into a matrix
room_type_percentages = room_type_percentages.pivot(index='neighbourhood', columns='room_type', values='percentage')

# plot the horizontal stacked bar chart
room_type_percentages.plot(kind='barh', stacked=True, figsize=(10,6))
plt.title('Top 10 Neighborhoods with the Highest Percentage of Each Room Type')
plt.xlabel('Percentage')
plt.ylabel('Neighborhood')
plt.show()


```

```python
# Chart - 11 visualization code
#17. What is the percentage of each room type available in each neighborhood?

# select the top 20 neighbourhoods by number of listings
top_neighbourhoods = airbnb_df['neighbourhood'].value_counts().head(20).index

# filter the data to only include the top 20 neighbourhoods
airbnb_top20 = airbnb_df[airbnb_df['neighbourhood'].isin(top_neighbourhoods)]

# calculate the percentage of each room type available in each neighborhood
room_type_percentages = airbnb_top20.groupby(['neighbourhood', 'room_type'])['id'].count().reset_index(name='count')
room_type_percentages['percentage'] = room_type_percentages.groupby('neighbourhood')['count'].apply(lambda x: x / x.sum() * 100)

# use pivot to reshape the data into a matrix
room_type_percentages = room_type_percentages.pivot(index='neighbourhood', columns='room_type', values='percentage')

# plot the stacked bar chart
room_type_percentages.plot(kind='bar', stacked=True, figsize=(15,10))
plt.title('Percentage of Each Room Type Available in Each Neighborhood (Top 20)')
plt.xlabel('Neighborhood')
plt.ylabel('Percentage')
plt.show()
```

# **7- HEATMAP**

```python
# Chart - 10 visualization code
#16. What is the busiest month for Airbnb in each neighborhood?

import seaborn as sns

# create a new column 'month' based on the 'last_review' column
airbnb_df['month'] = pd.DatetimeIndex(airbnb_df['last_review']).month

# group the data by neighborhood and month and count the number of listings
listings_by_month = airbnb_df.groupby(['neighbourhood', 'month']).size().reset_index(name='counts')

# use pivot to reshape the data into a matrix
listings_by_month = listings_by_month.pivot(index='neighbourhood', columns='month', values='counts')

# get the sum of listings in each neighborhood across all months
total_listings_by_neighborhood = listings_by_month.sum(axis=1)

# get the top 20 neighborhoods based on the number of listings
top_20_neighborhoods = total_listings_by_neighborhood.sort_values(ascending=False)[:20].index

# filter the data to only include the top 20 neighborhoods
listings_by_month_top_20 = listings_by_month.loc[top_20_neighborhoods]

# plot the heatmap
plt.figure(figsize=(15,10))
sns.heatmap(listings_by_month_top_20, cmap='YlGnBu', annot=True, fmt='g', linewidths=.5)
plt.title('Busiest Month for Airbnb in Each Neighborhood (Top 20)')
plt.xlabel('Month')
plt.ylabel('Neighborhood')
plt.show()
```
